```python
import pandas as pd
import numpy as np
```

### Download the data and load it to Pandas. 

You can find them [here](https://drive.google.com/file/d/1NY6cmF9Shjw-dD7BD6bNmfcIVz-kQcFR/view?usp=sharing).

```python
cast = pd.read_csv(
  './_data/Pandas_exercise/imdb_pandas/cast.csv',
  index_col=None)
cast.head()
```

```python
release_dates = pd.read_csv(
  './_data/Pandas_exercise/imdb_pandas/release_dates.csv',
  index_col=None,
  parse_dates=['date'])
release_dates.head()
```

### Count the number of movies with "Christmas" in their title for each month, that is released are the USA.

```python
release_dates['month'] = release_dates['date'].dt.month
```

```python
release_dates_christmas_usa = release_dates[
  (release_dates['title'].str.contains('Christmas')) &
  (release_dates['country'] == 'USA')]
release_dates_christmas_usa[['title', 'month']].groupby('month').count()
```

### Count the number of movies with "Christmas" in their title for each month, that are released in Canada.

```python
release_dates_christmas_canada = release_dates[
  (release_dates['title'].str.contains('Christmas')) &
  (release_dates['country'] == 'Canada')]
release_dates_christmas_canada[['title', 'month']].groupby('month').count()
```

### Count the number of movies in which the titles start with "The Hobbit" for each month, that are released in the USA.

```python
release_dates_hobbit_usa = release_dates[
  (release_dates['title'].str.match('^The Hobbit')) &
  (release_dates['country'] == 'USA')]
release_dates_hobbit_usa[['title', 'month']].groupby('month').count()
```

### Count the number of movies with "Romance" in their title for **each day of the week**, that are released in the USA.

```python
release_dates['day of week'] = release_dates['date'].dt.weekday
release_dates['day of week'].unique()
```

```python
release_dates_romance_usa = release_dates[
  (release_dates['title'].str.contains('Romance').fillna(False)) &
  (release_dates['country'] == 'USA')]
release_dates_romance_usa[['title', 'day of week']].groupby('day of week').count()
```

### Count the number of movies with "Action" in their title for **each day of the week**, that are released in the USA.

```python
release_dates_action_usa = release_dates[
  (release_dates['title'].str.contains('Action').fillna(False)) &
  (release_dates['country'] == 'USA')]
release_dates_action_usa[['title', 'day of week']].groupby('day of week').count()
```

### On which date was each Judi Dench movie from the 1990s released in the USA?

```python
cast['decade'] = (cast['year']/10).apply(np.floor).apply(int) * 10
cast
```

```python
cast_dates = pd.merge(cast, release_dates, on=['title', 'year'], how='left').dropna(subset=['date'])
cast_dates.head()
```

```python
cast_dates_judi_dench = cast_dates[
  (cast_dates['name'] == 'Judi Dench') &
  (cast_dates['decade'] == 1990) &
  (cast_dates['country'] == 'USA')]
cast_dates_judi_dench[['title', 'date']]
```

### In which months do films with the actress Judi Dench tend to be released in the USA?

```python
cast_dates_judi_dench[['month', 'n']].groupby('month').count()
```
